```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder,StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
le=LabelEncoder()
```

```python
cat_c=df.select_dtypes("object")
cat_c
```

```python
for i in cat_c:
    cat_c[i]=le.fit_transform(cat_c[i])
```

```python
cat_c
```

```python
num = df.select_dtypes(["int64","float64"])
```

```python
num
```

```python
num.describe()
```

```python
df_final=pd.concat([cat_c,num],axis=1)
df_final
```

```python
x=df_final.drop(columns=["satisfaction"])
y=df_final["satisfaction"]
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,train_size=0.7)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
log=LogisticRegression(max_iter=1000)
log.fit(xtrain,ytrain)
pred=log.predict(xtest)
```

```python
log.score(xtrain,ytrain)
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
accuracy_score(ytest,pred)
```

```python
print(classification_report(ytest,pred))
```

```python
print("Accuracy Score for test:  ",accuracy_score(ytest,pred))
print("Accuracy Score for Trian: ",accuracy_score(ytest,pred))
```

```python
from sklearn.metrics import roc_curve,roc_auc_score
```

```python
pred_prob=log.predict_proba(xtest)[:,1]
pred_prob
```

```python
fbr,tpr,threshold=roc_curve(ytest,pred_prob)
```

```python
auc_score=roc_auc_score(ytest,pred_prob)
```

```python
auc_score
```

```python
print(threshold)
print(fbr)
print(tpr)
```

```python
plt.plot(fbr,tpr,label=f"ROC-CURVE(AUC={auc_score:.2f})",color="purple")
plt.xlabel("False positive Rate")
plt.ylabel("True positive rate")
plt.title("ROC_AUC_CURVE")
plt.grid(True)
plt.show()
```
